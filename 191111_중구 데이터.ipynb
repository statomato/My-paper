- value -> y로 이름 바꾸기
- month, weekend 변수 만들기 
- temp_level, rain_level, wind_level, O3_level, dust_PM10_level, dust_PM2_level 변수 만들기


```python
def rain_level(rain) :
    if rain >= 30 :
        return 3
    else :
        if rain >=15 :
            return 2
        else :
            if rain >=3 :
                return 1
            else : return 0

def dust_PM2_level(dust) :
    if dust >= 76 :
        return 3
    else :
        if dust >=36 :
            return 2
        else :
            if dust >=16 :
                return 1
            else : return 0      

def dust_PM10_level(dust) :
    if dust >= 151 :
        return 3
    else :
        if dust >=81 :
            return 2
        else :
            if dust >=31 :
                return 1
            else : return 0

def wind_level(wind) :
    if wind >= 14 :
        return 3
    else :
        if wind >=9 :
            return 2
        else :
            if wind >=4 :
                return 1
            else : return 0

def O3_level(O3) :
    if O3 >= 0.151 :
        return 3
    else :
        if O3 >=0.091 :
            return 2
        else :
            if O3 >=0.031 :
                return 1
            else : return 0
            
car_jg['dust_PM10_level'] = car_jg['PM10'].apply(lambda x : dust_PM10_level(x))
car_jg['dust_PM2_level'] = car_jg['PM2.5'].apply(lambda x : dust_PM2_level(x))
car_jg['rain_level'] = car_jg['rain'].apply(lambda x : rain_level(x))       
car_jg['wind_level'] = car_jg['wind'].apply(lambda x : wind_level(x))            
car_jg['O3_level'] = car_jg['O3'].apply(lambda x : O3_level(x))

bins = [-40,-10,10,20,30,50]
bins_names = [0,1,2,3,4] # 구간별 이름
car_jg['temp_level'] = pd.cut(car_jg['temperature'], bins, labels=bins_names)
```

```python
car_jg.rename(columns = {'value' : 'y'}, inplace = True)
```

```python
car_jg['weekend'] = 0
car_jg['weekend'][car_jg['days'].isin(['토','일'])] = 1
```

```python
car_jg['month'] = car_jg.date.apply(lambda x : int(str(x)[-4:-2]))
car_jg['weekend'] = car_jg.days.apply(lambda x : 1 if (x == '토') | (x == '일') else 0)
```
