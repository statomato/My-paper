```python
import pandas as pd
car_df1 = pd.read_pickle('car_df1.pkl')
```

```python
link_id = pd.read_csv('link_id.csv',encoding='utf-8')
```

```python
link_id.head()
```

```python
link_id['start_lat']=link_id.start_location.str.split(',').str[0]
link_id['start_lon']=link_id.start_location.str.split(',').str[1]
link_id['end_lat']=link_id.end_location.str.split(',').str[0]
link_id['end_lon']=link_id.end_location.str.split(',').str[1]
```

```python
link_id.head()
```

```python
link_id.drop(['start_location','end_location'],axis=1,inplace=True)
```

```python
link_id.head()
```

```python
tt = link_id.drop(['road_name','start_point','end_point'],axis=1)
tt.head()
```

```python
car_df1 = pd.merge(car_df1, tt, on='link_id', how='inner')
```

```python
pd.set_option('display.max_columns', 500)
car_df1.head()
```

```python
car_df1.start_lat = car_df1.start_lat.astype(float)
```

```python
car_df1.start_lon = car_df1.start_lon.astype(float)
car_df1.end_lat = car_df1.end_lat.astype(float)
car_df1.end_lon = car_df1.end_lon.astype(float)
```

```python
car_df1.dtypes
```

```python
car_df1.to_csv('car_df1.csv',encoding='utf-8-sig')
```

```python
car_df1.to_pickle("car_df1.pkl")
```
